```python
import tensorflow as tf
import numpy as np
import copy
import math
import os

path = './NR-ER/\
NR-ER-train/names_onehots.npy'
label = np.genfromtxt('./NR-ER/\
NR-ER-train/names_labels.csv',delimiter=',',dtype=None)
data = np.load(path, allow_pickle = True)
m = np.atleast_1d(data)
dictionary = dict(m[0])
onehotArr = dictionary['onehots']
nameArr = dictionary['names']
print(dictionary)
dictlength = len(onehotArr[0])
datalength = len(onehotArr)
onelength = len(onehotArr[0][0])
print(dictlength, datalength, onelength)
```

```python
def add_layer(inputs, in_size, out_size, activation_function = None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus = tf.matmul(inputs, Weights) + biases

    if activation_function is None:
        outputs = Wx_plus
    else:
        outputs = activation_function(Wx_plus)
    return outputs

import matplotlib.pyplot as plt
x_data = onehotArr
c_data = label
#y_data = np.square(x_data) - 0.5 + noise

xs = tf.placeholder(tf.float32, [dictlength, onelength], name = 'x_input')
ys = tf.placeholder(tf.float32, [None, 1], name = 'y_input')

l1 = add_layer(xs, onelength, dictlength, activation_function = tf.nn.relu)
prediction = add_layer(l1, dictlength, 1, activation_function = None)

loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys - prediction),
                                reduction_indices = [1]))
## 0.1: learning rate
with tf.name_scope('train'):
    train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
sess = tf.Session()
writer = tf.summary.FileWriter('logs/', sess.graph)
init = tf.initialize_all_variables()
count1 = 0
count2 = 0
sess.run(init)
for i in range(datalength):
    #print(onehotArr.shape)
    #print(c_data[i][0])
    sess.run(train_step, feed_dict = {xs: x_data[i], ys: [[c_data[i][1]]]})
    prediction_value = sess.run(prediction, feed_dict = {xs: x_data[i]})
    if (c_data[i][1]==1 and abs(prediction_value[0])>0.5) or (c_data[i][1]==0 and abs(prediction_value[0])<0.5):
        if i<datalength/2:
            count1+=1
        else:
            count2+=1
    else:
        print('NOOK')
            # Why placeholder? x_data can be divided (not used in this program.)
print(count1, count2)
print(datalength)
```
